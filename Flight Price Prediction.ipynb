# Feature Engineering

1. Deriving features like day, month and year from field 'Date_of_Journey'. This is the first feature engineering procees.

```python
#Deriving features like day, month and year from field 'Date_of_Journey' and adding it as new fields in final_df.

final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
#Fields present in data frame final_df
final_df.info()
```

Here we will change the Dtype of the fields: Date, Month and year to int type.

```python
#Changing data type to int type of the following fields:
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
#Fields present in data frame final_df
final_df.info()
```

Now we will remove the field 'Date_of_Journey'

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head()
```

2. Deriving another feature from 'Arrival_Time' that is Arrival_Time_hr and Arrival_Time_min

Here, the Arrival_Time column we have some time those are combined with date, but we need only the time part so extracting the very first element.

```python
final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
#Arrival time hour part:
final_df['Arrival_Time_hour']=(final_df['Arrival_Time'].str.split(' ').str[0]).str.split(':').str[0]
#Since the value is in string converting it into int type.
final_df['Arrival_Time_hour']=final_df['Arrival_Time_hour'].astype(int)

#Arrival time minut part:
final_df['Arrival_Time_minute']=(final_df['Arrival_Time'].str.split(' ').str[0]).str.split(':').str[1]
#Since the value is in string converting it into int type.
final_df['Arrival_Time_minute']=final_df['Arrival_Time_minute'].astype(int)
```

```python
final_df['Arrival_Time_hour']
```

```python
final_df['Arrival_Time_minute']
```

```python
#Dropping field 'Arrival_Time' 
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

3. Deriving another feature from 'Dep_Time' that is Dep_Time_hr and Dep_Time_min

```python
#Departure time hour part:
final_df['Departure_Time_hour']=final_df['Dep_Time'].str.split(':').str[0]
#Since the value is in string converting it into int type.
final_df['Departure_Time_hour']=final_df['Departure_Time_hour'].astype(int)

#Departure time minut part:
final_df['Departure_Time_minute']=final_df['Dep_Time'].str.split(':').str[1]
#Since the value is in string converting it into int type.
final_df['Departure_Time_minute']=final_df['Departure_Time_minute'].astype(int)
```

```python
final_df['Departure_Time_hour']
```

```python
final_df['Departure_Time_minute']
```

```python
#Dropping field Dep_Time from final_df
final_df.drop('Dep_Time',axis=1,inplace=True)
```

4. Handling the field 'Total_Stops'.
 
In the dataframe final_df the field 'Total_Stops' is a categorical data type which tells number of stoppage between source and destination.

```python
final_df['Total_Stops'].unique()
```

Here, replacing the uniques values by following values:
   1. 'non-stop': 0
   2. '1 stops': 1
   3. '2 stop': 2
   4. '3 stops': 3
   5. '4 stops': 4
   6.  nan: 3, since usually for three stoppage it may take more than 20 hours of stoppage.

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, 'nan': 3})
```

```python
final_df.head()
```

```python
#Dropping field Route from final_df
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head()
```

5. Converting the duration in minutes:

```python
#Hour part of duration:
final_df['Duration_hour']=final_df['Duration'].str.split(' ').str[0].str[:-1]
final_df['Duration_hour']=final_df['Duration_hour'].astype(int)

#Minute part of duration:
final_df['Duration_minute']=final_df['Duration'].str.split(' ').str[1].str[:-1]
final_df['Duration_minute']=final_df['Duration_minute'].astype(int)
```

```python
final_df['Duration_hour']
```

```python
final_df['Duration_minute']
```

```python
#Converting the NaN part of minute to zero
final_df['Duration_minute']=final_df['Duration_minute'].fillna(0)
```

```python
final_df['Duration_minute'].isnull().sum()
```

There is no null record.

```python
final_df['Duration_minute']=final_df['Duration_minute'].astype(int)
```

```python
final_df['Duration_minute']
```

```python
final_df.info()
```

```python
final_df['Duration'] = final_df['Duration_hour']*60 + final_df['Duration_minute']
```

```python
final_df.head()
```

Now we will drop the two fields: Duration_hour and Duration_minute

```python
#Dropping field Duration_hour from final_df
final_df.drop('Duration_hour',axis=1,inplace=True)

#Dropping field Duration_minute from final_df
final_df.drop('Duration_minute',axis=1,inplace=True)
```

```python
final_df.info()
```

# Dealing with categorical data fields

Here following are the fields with categorical data type:
   1. Airline                 
   2. Source

   4. Additional_Info

So using scikit learn's label encoding feature for it.
By using LabelEncoder() we will categorise them into assigned integer.

# Using Label Encoding:

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df['Airline_LE']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source_LE']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination_LE']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info_LE']=labelencoder.fit_transform(final_df['Additional_Info'])
```
